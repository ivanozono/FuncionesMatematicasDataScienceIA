**Simple Linear Function**

---

```python
N = 100
m = -1
b = 3
def f(x):
  return m*x+b
x = np.linspace(-10,10, num=N)
y = f(x)
fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
```

This code represents the implementation and visualization of a simple linear function.

---

Code Explanation:

1 Variable Definitions: N = 100: This sets the number of data points that will be generated.
m = -1: This is the slope of the linear function. A negative slope indicates a line that decreases as 'x' increases.
b = 3: This is the y-intercept, meaning when x is 0, y will be 3.

2 Function Definition: Here, we define a linear function f(x) that takes in a value of x and returns the corresponding y-value based on the line equation y=mx+b.


3 Data Generation: Using the numpy function linspace, we generate an array of N equally spaced values of x ranging from -10 to 10.

4 Applying Function to Data: Here, we obtain the corresponding y-values for each x-value by applying our linear function f(x).

5 Visualization: First, we initialize a figure and axis object using plt.subplots().
Using ax.plot(x,y), we plot our generated data points (x,y) to visualize the line.
ax.grid() is employed to add gridlines, making the plot easier to interpret.

**Logarithmic Function**

---

```python
def f(x):
  return np.log2(x)
x = np.linspace(0.001,256, num=1000)
plt.plot(x,f(x))
```



This code delineates the construction and visualization of the base-2 logarithmic function.

---

**Code Explanation:**

1. **Function Definition:**
   
    The function `f(x)` is designed to compute the base-2 logarithm of the input value `x`.

2. **Data Generation:**
   
    With the numpy `linspace` method, we produce an array of 1000 equally spaced values of x in the range from 0.001 to 256. Starting from 0.001 prevents a mathematical error due to the logarithm of zero being undefined.

3. **Visualization:**
   
    By employing `plt.plot(x, f(x))`, we graphically represent our x-values against the computed y-values based on the logarithmic function. The resultant plot shows the characteristics of the base-2 logarithm curve, highlighting the logarithmic growth as `x` increases.

**Heaviside Step Function**

---

```python
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y
N=1000
x = np.linspace(-10,10, num=N)
y = H(x)
plt.plot(x,y)
```



This code illustrates the creation and visualization of the Heaviside step function, often referred to simply as the "step function".

---

**Code Explanation:**

1. **Function Definition:**

    The function `H(x)` models the step function, which has the following characteristics:
    - It returns 0 for all values of `x` less than 0.
    - It returns 1 for all values of `x` greater than or equal to 0.
    This behavior is achieved by:
    - First initializing a zero array `Y` with the same length as the input `x`.
    - Iterating over the input values using `enumerate` to both get the current value and its index.
    - If the current value is greater than or equal to 0, the function sets the corresponding value in `Y` to 1.

2. **Data Generation:**
   
    We generate `N=1000` equally spaced values of `x` ranging from -10 to 10 using numpy's `linspace` function.

3. **Function Application to Data:**
   
    The y-values are computed by passing the generated `x` values through the step function `H(x)`.

4. **Visualization:**
   
    The graph is plotted using `x` as the independent variable and `y` as the dependent variable. The resulting graph displays a sharp transition from 0 to 1 at the origin, reflecting the behavior of the step function.
