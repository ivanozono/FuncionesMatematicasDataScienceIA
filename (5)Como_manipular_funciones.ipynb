 **Manipulating Mathematical Functions**

---

 We will explore how to manipulate mathematical functions, a crucial skill in data science and artificial intelligence. We will cover vertical and horizontal shifts, stretching and compressing, and reflections of functions.

```python
import numpy as np
import matplotlib.pyplot as plt
```

---
**Vertical and Horizontal Shifts**

---

We can shift the graph of a function vertically or horizontally by adding or subtracting a constant 'c' to the function or its variable.

```python
# Define the function and the constant
def f(x): return np.sin(x)
c = 2

# Generate x values
x = np.linspace(-15, 15, num=1000)

# Apply the shifts to the function
y1 = f(x) + c  # Shift up by c
y2 = f(x) - c  # Shift down by c
y3 = f(x - c)  # Shift right by c
y4 = f(x + c)  # Shift left by c

# Plot the original function and the shifted functions
plt.figure(figsize=(12, 8))
plt.plot(x, f(x), label='Original function')
plt.plot(x, y1, label='Shift up by c')
plt.plot(x, y2, label='Shift down by c')
plt.plot(x, y3, label='Shift right by c')
plt.plot(x, y4, label='Shift left by c')

# Add grid, legend, and axis lines
plt.grid(True)
plt.legend()
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title('Vertical and Horizontal Shifts')
plt.show()
```



The code demonstrates how to manipulate a given mathematical function, specifically shifting it vertically and horizontally. The transformations are then visualized on a single plot.

---

**Code Explanation:**

1. **Function Definition and Constants Initialization:**
   
    The function \( f(x) \) is defined as the sine function. The constant \( c \) is initialized to the value 2.

2. **Generating Input Values:**

    Using numpy's `linspace`, a set of 1000 equally spaced values between -15 and 15 is generated for \( x \).

3. **Function Transformations:**
    
    - **Vertical Shifts:**
   
        Vertical shifts involve adding or subtracting a constant from the function's output values. `y1` represents the original function shifted up by `c` and `y2` represents it shifted down by `c`.

    - **Horizontal Shifts:**
      
        Horizontal shifts involve adding or subtracting a constant from the input values of the function. `y3` represents the original function shifted to the right by `c` and `y4` represents it shifted to the left by `c`.

4. **Visualization:**
  
    - The various transformed functions alongside the original are plotted on a single graph.
    - Gridlines are added for clarity.
    - A legend differentiates between the original and transformed functions.
    - The lines `plt.axhline(0, color='black',linewidth=0.5)` and `plt.axvline(0, color='black',linewidth=0.5)` are used to draw the x-axis and y-axis respectively.
    - A title 'Vertical and Horizontal Shifts' provides context to the visual representation.

**Stretching and Compressing Functions**

---



We can stretch or compress the graph of a function by multiplying the function or its variable by a constant 'c'. A stretch or compression is vertical if the function is multiplied by 'c' and horizontal if the variable is multiplied by 'c'.

```python
# Apply the stretches and compressions to the function
y1 = c * f(x)  # Vertical stretch by c
y2 = (1/c) * f(x)  # Vertical compression by c
y3 = f(c * x)  # Horizontal compression by c
y4 = f((1/c) * x)  # Horizontal stretch by c

# Plot the original function and the stretched/compressed functions
plt.figure(figsize=(12, 8))
plt.plot(x, f(x), label='Original function')
plt.plot(x, y1, label='Vertical stretch by c')
plt.plot(x, y2, label='Vertical compression by c')
plt.plot(x, y3, label='Horizontal compression by c')
plt.plot(x, y4, label='Horizontal stretch by c')

# Add grid, legend, and axis lines
plt.grid(True)
plt.legend()
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title('Stretching and Compressing')
plt.show()
```



This code illustrates how to apply vertical and horizontal stretches and compressions to a given mathematical function, using the sine function as an exemplar. The effects of these transformations are then visualized in one cohesive plot.

---

**Code Explanation:**

1. **Function Transformations:**
    
    - **Vertical Stretch and Compression:**
       
        To achieve a vertical stretch, the function's output values are multiplied by a factor greater than 1. Conversely, to achieve a vertical compression, the function's output values are multiplied by a factor between 0 and 1. Here, for a stretch, we multiply by `c` and for compression, we multiply by `1/c`.
    
    - **Horizontal Stretch and Compression:**
       
        Horizontal stretches and compressions are performed by manipulating the function's input values. For a horizontal compression, the input values are multiplied by a factor greater than 1. In contrast, for a horizontal stretch, they are multiplied by a factor between 0 and 1. Here, for compression, we multiply `x` by `c` and for stretch, we multiply `x` by `1/c`.

2. **Visualization:**
   
    The transformed functions, along with the original, are visualized on one plot:
    - The plotting function displays the original function and its transformations.
    - Gridlines and axis lines (x and y axis) are added to the plot for better clarity.
    - A legend distinguishes between the original and transformed functions.
    - A title, 'Stretching and Compressing', provides context to the visualization.

**Reflections of Functions**

---


We can reflect the graph of a function about the x-axis or y-axis by multiplying the function or its variable by -1. A reflection is about the x-axis if the function is multiplied by -1 and about the y-axis if the variable is multiplied by -1.

```python
# Apply the reflections to the function
y1 = -f(x)  # Reflection about the x-axis
y2 = f(-x)  # Reflection about the y-axis

# Plot the original function and the reflected functions
plt.figure(figsize=(12, 8))
plt.plot(x, f(x), label='Original function')
plt.plot(x, y1, label='Reflection about the x-axis')
plt.plot(x, y2, label='Reflection about the y-axis')

# Add grid, legend, and axis lines
plt.grid(True)
plt.legend()
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title('Reflections')
plt.show()
```



This code showcases how to implement vertical and horizontal reflections of a given mathematical function. The reflection transformations are subsequently visualized on a unified plot.

---

**Code Explanation:**

1. **Function Transformations:**
    
    - **Reflections:**
      
        A reflection about the x-axis inverts the output values of a function, which is achieved by taking the negative of the function's output. Similarly, a reflection about the y-axis switches the direction of the function along the x-axis. This is realized by taking the negative of the function's input values.

2. **Visualization:**
   
    - The transformations and the original function are jointly visualized on the plot.
    - A grid is overlayed on the plot to aid in the visualization of function behaviors.
    - A legend is used to distinguish the original function from its reflections.
    - The x and y axes are displayed using the `axhline` and `axvline` functions.
    - The title, 'Reflections', succinctly provides the overarching theme of the visualization.
